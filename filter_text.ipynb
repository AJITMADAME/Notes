```python
# Function to handle complex rows
def handle_complex_rows(data_lines, headers):
    row_buffer = []
    for line in data_lines:
        # Split the line into words
        words = re.split(r'\s{2,}|\t', line.strip())
        if len(words) >= len(headers):
            if row_buffer:
                processed_rows.append(row_buffer)
                row_buffer = []
            processed_rows.append(words[:len(headers)])
            if len(words) > len(headers):
                row_buffer = words[len(headers):]
        else:
            row_buffer.extend(words)
    if row_buffer:
        processed_rows.append(row_buffer)

handle_complex_rows(data_lines, headers)
# Save the cleaned DataFrame to Excel
output_excel_path = r"E:\text_ext\op1/extracted_table.xlsx"
df.to_excel(output_excel_path, index=False)
print(f"Extracted table saved to {output_excel_path}")
```

```python
import re
import pandas as pd

# Example text data
text = [
    ['8096/INF109K01049/ICICI Prudential Liquid Fund-Direct Plan -Growth'],
    ['Transaction Type', 'Date', 'Amount (INR)', 'NAV(INR)', 'Price (INR)', 'No.of Units', 'Balance Units'],
    ['Purchase -DIRECT', '363.0944.', '363.0944', '358015.712', '129993500.32', '18/06/2024', '7486356.490'],
    ['Gross Amount: Rs.130000000/-; Stamp Duty charges Levied', '6499.68'],
    ['Invalid Purchase 06-JUN-2024-ISIP Rejection-', '18/06/2024', 'Payment not received-'],
    ['47497625.12', '363.1644', '363.1644', 'Purchase DIRECT', '130788.219', '19/06/2024', '7617144.709'],
    ['2374.88', 'Gross Amount: Rs.47500000/-; Stamp Duty charges Levied'],
    ['Units Under Lien: 0.000', 'Name Of Financier: Not Applicable'],
    ['Free units available TO REDEEMTSUT TCH-7617144 700 1']
]

# Flatten and join text
flat_text = ' '.join([item for sublist in text for item in sublist])

# Define patterns to extract information
patterns = {
    'transaction_type': r'(Purchase -DIRECT|Purchase DIRECT|Redemption|Redemption DIRECT)',
    'date': r'\d{2}/\d{2}/\d{4}',
    'price': r'Price \(INR\):?\s*([\d,.]+)',
    'amount': r'Amount \(INR\):?\s*([\d,.]+)',
    'nav': r'NAV\(INR\):?\s*([\d,.]+)',
    'no_of_units': r'No\.of Units:?\s*([\d,.]+)'
}

# Extract data using regex
data = []
dates = re.findall(patterns['date'], flat_text)
transaction_types = re.findall(patterns['transaction_type'], flat_text)
prices = re.findall(patterns['price'], flat_text)
amounts = re.findall(patterns['amount'], flat_text)
navs = re.findall(patterns['nav'], flat_text)
no_of_units = re.findall(patterns['no_of_units'], flat_text)

# Determine the maximum number of entries for rows
max_len = max(len(dates), len(transaction_types), len(prices), len(amounts), len(navs), len(no_of_units))

# Fill missing values with empty strings to align lengths
dates += [''] * (max_len - len(dates))
transaction_types += [''] * (max_len - len(transaction_types))
prices += [''] * (max_len - len(prices))
amounts += [''] * (max_len - len(amounts))
navs += [''] * (max_len - len(navs))
no_of_units += [''] * (max_len - len(no_of_units))

# Create a list of dictionaries for each row
for i in range(max_len):
    data.append({
        'transaction_type': transaction_types[i],
        'date': dates[i],
        'price': prices[i],
        'amount': amounts[i],
        'nav': navs[i],
        'no_of_units': no_of_units[i]
    })

# Convert to DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
import re
import pandas as pd

# Example text data
text = [
    ['8096/INF109K01049/ICICI Prudential Liquid Fund-Direct Plan -Growth'],
    ['Transaction Type', 'Date', 'Amount (INR)', 'NAV(INR)', 'Price (INR)', 'No.of Units', 'Balance Units'],
    ['Purchase -DIRECT', '363.0944.', '363.0944', '358015.712', '129993500.32', '18/06/2024', '7486356.490'],
    ['Gross Amount: Rs.130000000/-; Stamp Duty charges Levied', '6499.68'],
    ['Invalid Purchase 06-JUN-2024-ISIP Rejection-', '18/06/2024', 'Payment not received-'],
    ['47497625.12', '363.1644', '363.1644', 'Purchase DIRECT', '130788.219', '19/06/2024', '7617144.709'],
    ['2374.88', 'Gross Amount: Rs.47500000/-; Stamp Duty charges Levied'],
    ['Units Under Lien: 0.000', 'Name Of Financier: Not Applicable'],
    ['Free units available TO REDEEMTSUT TCH-7617144 700 1']
]

# Flatten and join text
flat_text = ' '.join([item for sublist in text for item in sublist])

# Define patterns to match levied charges
pattern_before = r'([\d,.]+)[\s]*;[\s]*Stamp Duty charges Levied'
pattern_after = r'Stamp Duty charges Levied[\s\S]*?([\d,.]+)'

# Extract levied charges using regex
matches_before = re.findall(pattern_before, flat_text)
matches_after = re.findall(pattern_after, flat_text)

# Combine results and handle special cases where amounts might be directly before the phrase
levied_charges = matches_before + matches_after

# Remove potential duplicates and extra spaces
levied_charges = list(set(charge.strip() for charge in levied_charges))

# Print extracted levied charges for debugging
print("Extracted Levied Charges:")
print(levied_charges)

# Create DataFrame from extracted data
df = pd.DataFrame({'Levied Charges': levied_charges})
print("\nDataFrame:")
print(df)
```
