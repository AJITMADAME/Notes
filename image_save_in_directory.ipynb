```python
import os
import cv2

# Source directory path
source_directory = 'D:\\images'

# Create a destination directory if it doesn't exist
destination_parent_directory = 'D:\\output'
if not os.path.exists(destination_parent_directory):
    os.makedirs(destination_parent_directory)

# List all image files in the source directory
image_files = [f for f in os.listdir(source_directory) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Loop through each image file
for image_file in image_files:
    # Read the image
    image_path = os.path.join(source_directory, image_file)
    image = cv2.imread(image_path)

    # Perform OpenCV operations on the image (example: convert to grayscale)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Create a subdirectory in the destination directory for the current image
    image_name_without_extension = os.path.splitext(image_file)[0]
    destination_subdirectory = os.path.join(destination_parent_directory, image_name_without_extension)
    if not os.path.exists(destination_subdirectory):
        os.makedirs(destination_subdirectory)

    # Save the processed image in the subdirectory
    destination_path = os.path.join(destination_subdirectory, f'{image_name_without_extension}_processed.jpg')
    cv2.imwrite(destination_path, gray_image)
```

```python
import os
import cv2

# Source directory path
source_directory = 'input_images'

# Create a destination parent directory if it doesn't exist
destination_parent_directory = 'output_cropped_images_grouped'
if not os.path.exists(destination_parent_directory):
    os.makedirs(destination_parent_directory)

# List all image files in the source directory
image_files = [f for f in os.listdir(source_directory) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Define the number of images per group
images_per_group = 5

# Create groups of cropped images and save them
for group_index in range(0, len(image_files), images_per_group):
    # Create a subdirectory for the current group
    group_subdirectory = os.path.join(destination_parent_directory, f'group_{group_index // images_per_group}')
    if not os.path.exists(group_subdirectory):
        os.makedirs(group_subdirectory)

    # Loop through images in the current group
    for i in range(images_per_group):
        image_index = group_index + i
        if image_index < len(image_files):
            image_file = image_files[image_index]
            image_path = os.path.join(source_directory, image_file)
            image = cv2.imread(image_path)

            # Get image dimensions
            height, width, _ = image.shape

            # Crop the image into 5 parts (you can modify this part according to your needs)
            for j in range(5):
                crop_width = width // 5
                start_x = j * crop_width
                end_x = (j + 1) * crop_width
                cropped_image = image[:, start_x:end_x]

                # Save the cropped image in the subdirectory
                destination_path = os.path.join(group_subdirectory, f'{image_file}_part{j + 1}.jpg')
                cv2.imwrite(destination_path, cropped_image)
```
